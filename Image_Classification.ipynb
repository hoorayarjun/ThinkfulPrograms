```python
from PIL import Image
from numpy import asarray
pixel_data = []
ids = []
for i in range(1,6387):
    try:
        image = Image.open("GarageImages/image" + str(i) + ".jpg").convert('LA').resize((64,64), Image.ANTIALIAS)
        data = asarray(image)
        image.save("GarageImages/grayscale" +  str(i) + ".png")
        pixel_data.append(data[:,:,0].flatten())
        ids.append(i)
    except FileNotFoundError:
        print("No image for " + str(i))
```

```python
pixel_data
```

```python
import pandas as pd
df = pd.DataFrame(columns = ['ID'])
df['ID'] = ids
for j in range(0, 4096):
    transformed = []
    for i in pixel_data:
        transformed.append(i[j])
    column_name = "pixel"
    column_name += str(j)
    df[column_name] = transformed
df.head()
```

```python
labels_df = pd.read_csv('garage_door_labels.csv')
labels_df.head()
```

```python
training_set = pd.merge(labels_df, df, how="inner", on="ID")
training_set = training_set.sort_values("ID")
training_set.head()
```

```python
X = training_set.loc[:,'pixel0':'pixel4095']
Y = training_set['GarageDoorEntranceIndicator']
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(1000, 100, 100, 4), max_iter = 1000)
mlp.fit(X, Y)
mlp.score(X, Y)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(mlp, X, Y, cv=5)
```
