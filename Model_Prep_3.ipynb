```python
edu_no_time_df = edu_df.copy()
#This takes a long time, so as a proof of concept, I'll just use a couple
#columns = ['ENROLL', 'TOTAL_REVENUE', 'FEDERAL_REVENUE', 'STATE_REVENUE', 'LOCAL_REVENUE', 'TOTAL_EXPENDITURE', 'INSTRUCTION_EXPENDITURE', 'SUPPORT_SERVICES_EXPENDITURE', 'OTHER_EXPENDITURE', 'CAPITAL_OUTLAY_EXPENDITURE', 'GRADES_PK_G', 'GRADES_KG_G', 'GRADES_4_G', 'GRADES_8_G', 'GRADES_12_G', 'GRADES_1_8_G', 'GRADES_9_12_G', 'GRADES_ALL_G', 'AVG_MATH_4_SCORE', 'AVG_MATH_8_SCORE', 'AVG_READING_4_SCORE', 'AVG_READING_8_SCORE']
columns = ['ENROLL', 'FEDERAL_REVENUE', 'AVG_READING_4_SCORE']
for x in range(len(edu_no_time_df['STATE'])):
    for column in columns:
        if(pd.isnull(edu_no_time_df[column][x])):
            state = edu_no_time_df['STATE'][x]
            edu_no_time_df[column][x] = np.mean(edu_no_time_df.loc[edu_no_time_df['STATE'] == state][column])
edu_no_time_df
```

```python
edu_time_df = edu_df.copy()
columns = ['ENROLL']
for x in range(len(edu_time_df['STATE'])):
    for column in columns:
        if(pd.isnull(edu_time_df[column][x])):
            lower_val = np.NaN
            for i in range(x, -1, -1):
                if(edu_time_df['STATE'][x] == edu_time_df['STATE'][i]):
                    if(pd.notnull(edu_time_df[column][i])):
                        lower_val = edu_time_df[column][i]
                        i = -1
            higher_val = np.NaN
            for j in range(x, len(edu_time_df['STATE'])):
                if(edu_time_df['STATE'][x] == edu_time_df['STATE'][j]):
                    if(pd.notnull(edu_time_df[column][j])):
                        higher_val = edu_time_df[column][j]
                        j = len(edu_time_df['STATE'])
            if(pd.isnull(lower_val)):
                lower_val = higher_val
            if(pd.isnull(higher_val)):
                higher_val = lower_val
            edu_time_df[column][x] = np.mean([lower_val, higher_val])
edu_time_df
```
