```python
from sklearn import tree
import pandas as pd
from IPython.display import Image
import numpy as np

golf_df = pd.DataFrame()
golf_df['Outlook'] = ['sunny', 'sunny', 'overcast', 'rainy', 'rainy', 'rainy', 
                     'overcast', 'sunny', 'sunny', 'rainy', 'sunny', 'overcast',
                     'overcast', 'rainy']
golf_df['Temperature'] = [92, 86, 83, 50, 60, 53, 62,
                         75, 57, 72, 78, 69, 61, 71]
golf_df['Humidity'] = [40.3, 63.1, 35.3, 86.3, 72.4, 22.7, 27.0,
                      80.7, 15.4, 25.1, 18.6, 58.3, 47.2, 39.3]
golf_df['Play'] = ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 
                   'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']

golf_dat = pd.get_dummies(golf_df)
golf_dat
```

```python
from sklearn.model_selection import cross_val_score
import time

X = golf_dat[['Temperature', 'Humidity', 'Outlook_overcast', 'Outlook_rainy', 'Outlook_sunny']]
Y = golf_dat['Play_yes']

start_time = time.time()
clf = tree.DecisionTreeClassifier()
print(np.mean(cross_val_score(clf, X, Y, cv=5)))
print("{} seconds".format(time.time() - start_time))
```

```python
from sklearn import ensemble

start_time = time.time()
rfc = ensemble.RandomForestClassifier(n_estimators = 20)
print(np.mean(cross_val_score(rfc, X, Y, cv=5)))
print("{} seconds".format(time.time() - start_time))
```
