```python
import pandas as pd
import numpy as np

amazon_df = pd.read_fwf('amazon_cells_labelled.txt')
amazon_df
```

```python
split_text = amazon_df['Raw_Text'].str.split("\t")
reviews = []
sentiments = []
for i in range(len(split_text)):
    reviews.append(split_text[i][0])
    sentiments.append(int(split_text[i][1]))
amazon_df['Review'] = reviews
amazon_df['Sentiment'] = sentiments
amazon_df
```

```python
good_reviews = amazon_df.loc[amazon_df['Sentiment'] == 1]['Review']
word_dict = {}
for review in good_reviews:
    words = review.split(" ")
    for word in set(words):
        if word in word_dict.keys():
            word_dict[word] = word_dict.get(word) + 1
        else:
            word_dict.update({word: 1})
sorted_word_dict = [(k, word_dict[k]) for k in sorted(word_dict, key=word_dict.get, reverse=True)]
print(sorted_word_dict)
```

```python
bad_reviews = amazon_df.loc[amazon_df['Sentiment'] == 0]['Review']
word_dict = {}
for review in bad_reviews:
    words = review.split(" ")
    for word in set(words):
        if word in word_dict.keys():
            word_dict[word] = word_dict.get(word) + 1
        else:
            word_dict.update({word: 1})
sorted_word_dict = [(k, word_dict[k]) for k in sorted(word_dict, key=word_dict.get, reverse=True)]
print(sorted_word_dict)
```

```python
keyword1 = ['good', 'great', 'excellent', 'not good', 'best', 'disappoint', 'fine']
keyword2 = ['bad', 'terrible', 'waste', 'disappoint', 'unsatisfactory', 'first', 'perfect']
```

```python
keywords = ['good', 'great', 'excellent', 'bad', 'terrible', 'waste', 'not good', 'best', 'disappointed', 'disappoint', 'fine', 'unsatisfactory', 'quality', 'recommend', 'happy', 'first', 'the']

for key in keyword1:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    amazon_df[str(key)] = amazon_df.Review.str.contains(str(key),case=False)
for key in keywords:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    amazon_df[str(key)] = amazon_df.Review.str.contains(str(key),case=False)

amazon_df['allcaps'] = amazon_df.Review.str.isupper()
amazon_df['Positive Review'] = (amazon_df['Sentiment'] == 1)
amazon_df.loc[amazon_df['not good'] == True]
```

```python
data = amazon_df[['good', 'great', 'excellent', 'bad',
       'terrible', 'waste', 'allcaps','not good', 'best', 'disappointed', 'disappoint', 'fine', 'unsatisfactory', 'first']]
target = amazon_df['Positive Review']
# Our data is binary / boolean, so we're importing the Bernoulli classifier.
from sklearn.naive_bayes import BernoulliNB

# Instantiate our model and store it in a new variable.
bnb = BernoulliNB()

# Fit our model to the data.
bnb.fit(data, target)

# Classify, storing the result in a new variable.
y_pred = bnb.predict(data)

# Display our results.
print("Number of mislabeled points out of a total {} points : {}".format(
    data.shape[0],
    (target != y_pred).sum()
))
```

```python
imdb_file = open('imdb_labelled.txt', 'r')
imdb_text = imdb_file.read()
imdb_text = imdb_text.split("\n")
reviews = []
sentiments = []
for text in imdb_text:
    splt = text.split('\t')
    reviews.append(splt[0])
    sentiments.append(splt[1])
imdf = pd.DataFrame()
imdf['Review'] = reviews
imdf['Sentiment'] = sentiments
imdf.head()
```

```python
keywords = ['good', 'great', 'excellent', 'bad', 'terrible', 'waste', 'not good', 'best', 'disappointed', 'disappoint', 'fine', 'unsatisfactory', 'quality', 'recommend', 'happy', 'first', 'the']

for key in keywords:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    imdf[str(key)] = imdf.Review.str.contains(str(key),case=False)

for key in keyword2:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    imdf[str(key)] = imdf.Review.str.contains(str(key),case=False)
    
imdf['allcaps'] = imdf.Review.str.isupper()
imdf['Positive Review'] = (imdf['Sentiment'] == '1')
imdf.head()
```

```python
data2 = imdf[['good', 'great', 'excellent',
       'terrible', 'waste', 'allcaps','not good', 'best', 'disappointed', 'disappoint', 'fine', 'unsatisfactory', 'first']]
target2 = imdf['Positive Review']
# Our data is binary / boolean, so we're importing the Bernoulli classifier.

# Instantiate our model and store it in a new variable.
bnb2 = BernoulliNB()

# Fit our model to the data.
bnb2.fit(data2, target2)

# Classify, storing the result in a new variable.
y_pred2 = bnb2.predict(data2)

# Display our results.
print("Number of mislabeled points out of a total {} points : {}".format(
    data2.shape[0],
    (target2 != y_pred2).sum()
))
```
