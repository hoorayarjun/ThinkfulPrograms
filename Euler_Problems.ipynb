```python
def is_palindrome(x):
    x = str(x)
    if(len(x) <= 1):
        return True
    elif(x[0:1] == x[len(x)-1:len(x)]):
        return is_palindrome(x[1:len(x)-1])
    else:
        return False
max_palindrome = 0
for i in range(1000):
    factor_one = 999 - i
    for j in range(factor_one):
        factor_two = 999 - j
        val = factor_one * factor_two
        #print('{} , {}: {}, {}'.format(factor_one, factor_two, val, is_palindrome(val)))
        if(is_palindrome(val)):
            if(val>max_palindrome):
                max_palindrome = val
            break
print(max_palindrome)
```

```python
def prime_factors(x):
    output = []
    divisor = 2
    while(x >= 2):
        if(x%divisor == 0):
            output.append(divisor)
            x = x/divisor
            divisor = 2
        else:
            divisor += 1
        if(x == 2):
            output.append(2)
            return output
    return output
def least_multiple(x):
    factors = []
    for i in range(1,x):
        i_factors = prime_factors(i)
        dummy_factors = factors.copy()
        for fact in i_factors:
            if fact in dummy_factors:
                dummy_factors.remove(fact)
            else:
                factors.append(fact)
    print(factors)
    output = 1
    for fact in factors:
        output *= fact
    return output
print(least_multiple(20))
```
